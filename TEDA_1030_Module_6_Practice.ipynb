```python
# Question 11: The Company column contains errors. What kind of errors are they?
  # Sometimes the company names are capitalized, and sometimes they aren't, causing standardization issues.

df['Company'].unique()
```

```python
# Question 12: Fix the error shown above using the method of your preference. You may need to use ChatGPT and/or Google to figure out how to do this (hint: use the .str.title() method).
  # How many computers in this data set are from the company Apple?
    # 19 of the computers in the data set are from Apple

df['Company'] = df['Company'].replace({
  'acer': 'Acer',
  'hp': 'HP',
  'apple': 'Apple',
  'dell': 'Dell',
  'lenovo': 'Lenovo',
  'asus': 'Asus',
  'microsoft': 'Microsoft',
  'msi': 'MSI',
  'toshiba': 'Toshiba',
  'mediacom': 'Mediacom',
  'razer': 'Razer',
  'samsung': 'Samsung'
})
df['Company'].unique()

apple_comp = df[df['Company'] == 'Apple']
apple_comp.shape[0]
```

```python
# Question 13: It's difficult to tell, but some of the laptops seem to have extremely high prices.
  # Use the z-score method with a threshold of 3 to create a filter and determine which laptops are outliers in the data set in terms of price.
  # Using the outlier detection method from above, how many outliers exist in this data set according to price?
    # Using this method, 8 outliers exist.
      # My code returns 8, but that isn't an option on the quiz - 9 is the closest (all options are 1, 2, 5, 9).

average_price = df['Price'].mean()
std_dev_price = df['Price'].std()
df['Z-Score'] = (df['Price'] - average_price) / std_dev_price
df.loc[df['Z-Score'] > 3]
```
