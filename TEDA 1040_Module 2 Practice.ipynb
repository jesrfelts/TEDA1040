```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
df.info()
```

```python
# Question 1: Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state. 

plt.figure(figsize=(12, 2), dpi=100)
sns.countplot(x='state', data=df)
plt.show()
```

```python
# Question 2: Which state are most of the companies from?
    # California
```

```python
# Question 3: Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.

industry_jobs = df.groupby('Industry', as_index=False)['job_openings'].sum()
plt.figure(figsize=(8, 4), dpi=80)
sns.barplot(data = industry_jobs, x = 'Industry', y = 'job_openings')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Question 4: Which industry has the most job openings?
    # Fintech
```

```python
# Question 5: Using matplotlib/seaborn, create a line chart that shows the average growth rate for each founded year.

ave_growth = df.groupby('founded', as_index=False)['growth_percentage'].mean() 
plt.figure(figsize=(8, 4), dpi=80) 
sns.lineplot(data = ave_growth, x = 'founded', y = 'growth_percentage') 
plt.show()
```

```python
# Question 6: From the years 2000 to 2020, does the growth percentage seem to be increasing, decreasing, or not changing at all?
    # There are many dips, but overall it's increasing
```

```python
# Question 7: Using matplotlib/seaborn, create a scatter plot that compares the number of employees to estimated revenues for each company.

plt.figure(figsize=(8, 4), dpi=80) 
sns.scatterplot(data = df, x = 'employees', y = 'estimated_revenues') 
plt.show()
```

```python
# Question 8: What is the relationship between number of employees and estimated revenues?
    # As the number of employees increases, so does estimated revenues
```

```python
# Question 9: Using matplotlib/seaborn, create a histogram that measures the number of companies founded in each year.

plt.figure(figsize=(8, 4), dpi=80) 
sns.histplot(data = df, x = 'founded') 
plt.show()
```
